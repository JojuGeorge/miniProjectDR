```python

# -*- coding:utf-8 -*-
# Imports
import sys
import os
from cv2 import *
from Tkinter import * 
from Tkinter import Tk, Menu, Canvas
from PIL import Image, ImageTk
import tkFileDialog as filedialog
import test07_gui00


imgname = "test_dataset/11085_left.jpeg"
width = 500
height = 800




# Function definitions
def deleteImage(canvas):
    canvas.delete("all")
    return


def quitProgram():
    gui.destroy()
    

# Main window
gui = Tk()
# setting the size of the window
canvas = Canvas(gui, width = 700, height = 600)  
canvas.pack()  


def browsefunc():
    global imgname
    imgname = filedialog.askopenfilename()
    pathlabel.config(text=imgname,fg = "blue", font = "Times")
    
pathlabel = Label()
pathlabel.pack()


#RUNNING PROGRAM
def run():
    test07_gui00.reading(imgname)
    
    
    #showing image
    w = 430
    h = 320
    im = Image.open("img")
    image = im.resize((w,h))
    img = ImageTk.PhotoImage(image, master = gui)  
    canvas.create_image(10, 10, anchor=NW, image=img) 


    
def result():
    pathlabel.config(text=imgname,fg = "blue", font = "Times")
    pathlabel = Label()
    pathlabel.pack()
  
#PLACING BROWSE BUTTON
browsebutton = Button(gui, text="Browse",command=browsefunc)
browsebutton.pack()

#PREVIEW IMAGE BUTTON
#PreviewButton = Button(gui, text="Preview",command=loadImage)
#PreviewButton.pack()

#PROCEED BUTTON
ProceedButton = Button(gui, text="Proceed",command=run)
ProceedButton.pack()

#display count button
ResultButton = Button(gui, text="result",command=result)
ResultButton.pack()



gui.title("DR")
gui.mainloop()
```
