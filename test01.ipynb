```python
import cv2 as cv2   #importing opevcv
import numpy as np


symp_im = cv2.imread('symptoms/11085_left.jpeg');    #reading an image
grayScale_symp_im = cv2.cvtColor(symp_im, cv2.COLOR_BGR2GRAY);  #converting the orginal image to GRAY SCALE


#RESIZE IMAGE
symp_im_rs = cv2.resize(symp_im, (430, 320));     #orginal image resizing
gray_symp_im_rs = cv2.resize(grayScale_symp_im, (430, 320));    #grayscale image resizing

#SHOWING IMAGE
cv2.imshow('symptoms image',symp_im_rs);    #orginal image
cv2.imshow('Gray Scale - symptoms image',gray_symp_im_rs);    #gryascale image


#BINARISING & THRESHOLD SLIDER
#threshold slider code
#nothing() is called each time threshold value is updated
def nothing(x):
    if s1 == 1:
        binarise_type = cv2.THRESH_BINARY
    if s2 == 1:
        binarise_type = cv2.THRESH_BINARY_INV
    if s3 == 1:
        binarise_type = cv2.THRESH_OTSU
    if s4 == 1:
        binarise_type = cv2.THRESH_TRUNC
    if s5 == 1:
        binarise_type = cv2.THRESH_TOZERO
    if s6 == 1:
        binarise_type = cv2.THRESH_TOZERO_INV
        
    try:
        ret,thresh_symp_im = cv2.threshold(grayScale_symp_im, thresh_min, thresh_max, binarise_type);   #binarising based on threshold slider
        bin_symp_im_rs = cv2.resize(thresh_symp_im, (430, 320));    #Binary image resizing
        cv2.imshow('Binarised', bin_symp_im_rs);    #show binary image
    except UnboundLocalError:
        print 'UnboundLocalError: local variable "binarise_type" referenced before assignment'
        

    
cv2.namedWindow('Binarised')   #creates window
"""creates a trackbar(slider name, window to be displayed, threshold min, threshold max, 
   callback function which is executed everytime trackbar value changes)"""
cv2.createTrackbar("Thresh_MIN", "Binarised", 0, 255, nothing); 
cv2.createTrackbar("Thresh_MAX", "Binarised", 0, 255, nothing); 


#creating a switch
switch1 = 'THRESH_BINARY \n 0 : OFF 1 : ON'
switch2 = '\nTHRESH_BINARY_INV \n 0 : OFF 1 : ON'
switch3 = 'THRESH_OTSU \n 0 : OFF 1 : ON'
switch4 = '\nTHRESH_TRUNC \n 0 : OFF 1 : ON'
switch5 = 'THRESH_TOZERO \n 0 : OFF 1 : ON'
switch6 = '\nTHRESH_TOZERO_INV \n 0 : OFF 1 : ON'
cv2.createTrackbar(switch1, 'Binarised',0,1,nothing)
cv2.createTrackbar(switch2, 'Binarised',0,1,nothing)
cv2.createTrackbar(switch3, 'Binarised',0,1,nothing)
cv2.createTrackbar(switch4, 'Binarised',0,1,nothing)
cv2.createTrackbar(switch5, 'Binarised',0,1,nothing)
cv2.createTrackbar(switch6, 'Binarised',0,1,nothing)

    
#while loop to update the threshold value 'thresh'
while(1):
    k = cv2.waitKey(1) & 0xFF;
    if k == 27:
        break;
    thresh_min = cv2.getTrackbarPos('Thresh_MIN', 'Binarised'); #gets current trackbar value
    thresh_max = cv2.getTrackbarPos('Thresh_MAX', 'Binarised'); 

    #each time trachbar is changed the nothing func is called from the cv2.createTrackbar()
    
    s1 = cv2.getTrackbarPos(switch1, 'Binarised');
    s2 = cv2.getTrackbarPos(switch2, 'Binarised');
    s3 = cv2.getTrackbarPos(switch3, 'Binarised');
    s4 = cv2.getTrackbarPos(switch4, 'Binarised');
    s5 = cv2.getTrackbarPos(switch5, 'Binarised');
    s6 = cv2.getTrackbarPos(switch6, 'Binarised');
    
#END BINARISING & THRESHOLD SLIDER CODE :)



#keyboard binding function
k = cv2.waitKey(0);   
if k == 27:           #wait for ESC key to exit
    cv2.destroyAllWindows();     #destroys all windows created
```
